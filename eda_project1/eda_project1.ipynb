```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fill missing values in categorical columns with the mode (most frequent value)
categorical_cols = ["Weather_Conditions", "Type_of_Junction"]
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace = True)

# Fill missing values in numericalcolumns with the median
numerical_cols = ["Speed_Limit","Road_Width","Vehicle_Speed","Time_Taken_for_Emergency_Response"]
for  col in numerical_cols:
    df[col].fillna(df[col].median(),inplace = True)

# Verify if missing values are handled
df.isnull().sum()
```

```python
# Replace space with underscore for each column name
df.columns = df.columns.str.replace(' ','_')
print(df.columns)
# Change column values to Title case
df = df.apply(lambda x:x.str.upper()if x.dtype == object else x)
print(df)
```

```python
# check for duplicates
print("Duplicate count:")
print(df.duplicated().sum())
```

```python
# convert Datetime Format
import pandas as pd
df['Time_of_Day'] = pd.to_datetime(df['Time_of_Day'])
df.tail(5)
```

```python
# Round the float value in 2 Decimal point
df = df.round({'Distance_to_Nearest_Hospital':2, 'Distance_to_Nearest_Police_Station':2, 'Visibility':2,'Road_Width':2,'Road_Width':2,'Road_Surface_Friction_Coefficient':2,'Time_Taken_for_Emergency_Response':2})
```

```python
#Calculate correlation
correlation_matrix = df.corr()
print(correlation_matrix)

# Heatmap plot of  correlation
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Accident Data")
plt.show()
```

```python
# 1.  What is the most common accident type?
df['Type_of_Accident'].value_counts().idxmax()

plt.figure(figsize=(12,6))
sns.countplot(y='Type_of_Accident', data=df, order=df['Type_of_Accident'].value_counts().index, palette="mako")

plt.title('Most Common Types of Accidents', fontsize=14)
plt.xlabel('Count of Accidents', fontsize=12)
plt.ylabel('Type of Accident', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()

```

```python
# 10.  How to plot the number of accidents per day of the week?
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
accidents_per_day = df['Day_of_Week'].value_counts().reindex(
    ['MONDAY', 'TUESDAY', 'WEDNESDAY', 'THURSDAY', 'FRIDAY', 'SATURDAY', 'SUNDAY']
)

# Plotting
plt.figure(figsize=(8, 5))
accidents_per_day.plot(kind='bar', color='cornflowerblue', edgecolor='black')

# Labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.title('Number of Accidents per Day of the Week')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()

# value_counts() counts occurrences of each day.
#reindex([...]) ensures the days appear in order.
#plot(kind='bar') generates a bar chart.
#plt.grid(axis='y', linestyle='--', alpha=0.7) improves readability
```

```python
# 11 .  Distribution of Accidents by Weather Conditions (Pie Chart)
df['Weather_Conditions'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', figsize=(8, 8), colormap='jet'
)
plt.ylabel('')
plt.title('Accident Distribution by Weather Conditions')
plt.show()
```

```python
# 12 .  Number of Accidents per State (Bar Chart)
df['State'].value_counts().nlargest(10).plot(
    kind='bar', figsize=(10, 5), color='coral', edgecolor='black'
)
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.title('Top 10 States with the Most Accidents')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 13 .  Speed Limit vs. Number of Accidents (Histogram)
df['Speed_Limit'].hist(bins=20, color='orange', edgecolor='black', figsize=(8, 5))
plt.xlabel('Speed Limit')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents by Speed Limit')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 14 .  Time of Day vs. Number of Accidents (Line Chart)
df['Hour'] = pd.to_datetime(df['Time_of_Day']).dt.hour
accidents_per_hour = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
accidents_per_hour.plot(kind='line', marker='o', color='purple', linestyle='-')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.title('Accidents Trend Over Time of the Day')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.xticks(range(0, 24))
plt.show()
```

```python
# 15.  How does the number of accidents vary by light conditions?
plt.figure(figsize=(12,6))
sns.countplot(y='Light_Conditions', data=df, order=df['Light_Conditions'].value_counts().index, palette="plasma")

plt.title('Number of Accidents by Light Conditions', fontsize=14)
plt.xlabel('Count of Accidents', fontsize=12)
plt.ylabel('Light Conditions', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
# 16 .  Number of Accidents at Different Junction Types (Bar Chart)
df['Type_of_Junction'].value_counts().plot(
    kind='bar', figsize=(8, 5), color='blueviolet', edgecolor='black'
)
plt.xlabel('Type of Junction')
plt.ylabel('Number of Accidents')
plt.title('Accidents at Different Junction Types')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 17 .  Road Width vs. Number of Accidents (Histogram)
df['Road_Width'].hist(bins=15, color='teal', edgecolor='black', figsize=(8, 5))
plt.xlabel('Road Width')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents by Road Width')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 18 .  What is the percentage of accidents based on road conditions?
df['Road_Conditions'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', figsize=(8, 8), colormap='tab10', startangle=270
)
plt.ylabel('')
plt.title('Accidents by Road Conditions')
plt.show()

# This chart displays the proportion of accidents based on road conditions (Dry, Wet, Muddy, Snow-covered, etc.).
```
